# Base de dados

- Kiva Microfunds https://www.kiva.org/

```python
from pycaret.datasets import get_data
dataset = get_data('kiva')
```

```python
dataset.shape
```

```python
data = dataset.sample(1000, random_state=1).reset_index(drop=True)
data.shape
```

# Configuração do ambiente

```python
from pycaret.nlp import setup
experimento = setup(data = data, target = 'en', session_id = 123)
```

# Criação do modelo (topic model)

```python
from pycaret.nlp import models
models()
```

```python
from pycaret.nlp import create_model
lda = create_model('lda')
```

```python
print(lda)
```

```python
lda2 = create_model('lda', num_topics = 6, multi_core = True)
print(lda2)
```

# Visualização e previsão

```python
from pycaret.nlp import assign_model
lda_results = assign_model(lda)
lda_results
```

```python
from pycaret.nlp import plot_model
plot_model(plot = 'bigram')
```

```python
plot_model(lda, plot = 'frequency', topic_num = 'Topic 1')
```

```python
plot_model(lda, plot = 'frequency', topic_num = 'Topic 2')
```

```python
plot_model(lda, plot = 'topic_distribution')
```

```python
# T-distributed Stochastic Neighbor Embedding
plot_model(lda, plot = 'tsne')
```

```python
from pycaret.nlp import evaluate_model
evaluate_model(lda)
```

# Salvar e carregar o modelo

```python
from pycaret.nlp import save_model
save_model(lda,'Final Lda')
```

```python
from pycaret.nlp import load_model
saved_lda = load_model('Final Lda')
```
