# Base de dados

- Mice Protein Expression: níveis de expressão de 77 proteínas/modificações de proteínas que produziram sinais detectáveis na fração nuclear do córtex
- Mais informações: https://archive.ics.uci.edu/ml/datasets/Mice+Protein+Expression

```python
from pycaret.datasets import get_data
dataset = get_data('mice')
```

```python
dataset.shape
```

```python
data = dataset.sample(frac=0.95, random_state=1)
data.shape
```

```python
data_unseen = dataset.drop(data.index)
data_unseen.shape
```

```python
data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)
```

# Configuração do ambiente

```python
from pycaret.anomaly import setup
experimento = setup(data, normalize = True, ignore_features = ['MouseID'], session_id = 123)
```

# Criação do modelo

```python
from pycaret.anomaly import models
models()
```

```python
from pycaret.anomaly import create_model
iforest = create_model('iforest') # Isolation Forest
```

```python
print(iforest)
```

```python
from pycaret.anomaly import assign_model
iforest_results = assign_model(iforest)
iforest_results
```

# Visualização e previsão

```python
from pycaret.anomaly import plot_model
plot_model(iforest)
```

```python
from pycaret.anomaly import predict_model
unseen_predictions = predict_model(iforest, data=data_unseen)
unseen_predictions.head()
```

# Salvar e carregar o modelo

```python
from pycaret.anomaly import save_model
save_model(iforest,'Final IForest')
```

```python
from pycaret.anomaly import load_model
saved_iforest = load_model('Final IForest')
```

```python
new_prediction = predict_model(saved_iforest, data=data_unseen)
new_prediction
```

```python
outliers = new_prediction[new_prediction['Anomaly'] == 1]
outliers
```

```python
outliers.to_csv('outliers.csv')
```
